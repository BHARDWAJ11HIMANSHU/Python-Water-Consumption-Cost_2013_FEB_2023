Comparing 'Estimated' and 'Bill_Analyzed' values. Are there any patterns in their combinations?

```python
# Creating a cross-tabulation
cross_tab = pd.crosstab(df['Estimated'], df['Bill_Analyzed'], margins=True, margins_name='Total')

# Creating a heatmap to visualize the cross-tabulation
plt.figure(figsize=(10, 5))
sns.heatmap(cross_tab, annot=True, cmap="YlGnBu", fmt='d', cbar=True)

plt.title('Comparison of Estimated and Bill Analyzed Values')
plt.xlabel('Bill Analyzed')
plt.ylabel('Estimated')

plt.show()
```

Creating a heatmap to explore the correlations between numerical columns like 'Current_Charges' and 'Consumption_(HCF)'.

```python
numerical_columns = df[['Current_Charges', 'Consumption_(HCF)']]

# Calculating the correlation matrix
correlation_matrix = numerical_columns.corr()

plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlation Heatmap')
plt.show()
```

Comparing 'Meter_AMR' and 'Funding_Source'  to see if there is a relationship.


```python
cross_tab = pd.crosstab(df['Meter_AMR'], df['Funding_Source'], margins=True, margins_name='Total')

plt.figure(figsize=(12, 6))
ax=cross_tab.drop('Total', axis=1).plot(kind='bar', stacked=True, colormap='tab20')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='center')

plt.title('Relationship Between Meter AMR and Funding Source')
plt.xlabel('Meter AMR')
plt.ylabel('Count')
plt.xticks(rotation=0)

plt.legend(title='Funding Source', loc='center',prop={'size': 8})


plt.show()
```

Conducting a time series analysis to understand how 'Current_Charges' and 'Consumption_(HCF)' change over 'Revenue_Month'.

```python
# Setting up the  'Revenue_Month' as the index
df.set_index('Revenue_Month', inplace=True)

# here ,Resampling  the data to a monthly frequency and calculatimg the mean for each month
monthly_data = df.resample('M').mean()


plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['Current_Charges'], label='Current Charges', marker='o')
plt.plot(monthly_data.index, monthly_data['Consumption_(HCF)'], label='Consumption (HCF)', marker='o')

plt.title('Time Series Analysis of Current Charges and Consumption (HCF) Over Time')
plt.xlabel('Revenue Month')
plt.ylabel('Average Value')

plt.grid(True)
plt.legend()

plt.show()
```
